```r
p <- function(n){
    return(1-(factorial(n)*choose(365,n)/365^n))
}
```

```r
n <- seq(1,100,1)
df <- data.frame("n"=n, "p(n)"=p(n))
```

```r
df
```

```r
library('ggplot2')

plot <- ggplot(data=df, aes(x=n, y=p.n.)) +
        geom_line()
```

```r
plot + geom_hline(yintercept=0.50, color="red")
```
